```python
import time
import matplotlib.pyplot as plt
import numpy as np

from mnist import MNIST
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import check_random_state
```

```python
mndata = MNIST('mnist/')

X_train, y_train = mndata.load_training()
X_test, y_test = mndata.load_testing()

y_train = np.reshape(y_train, (training_samples, ))
y_test = np.reshape(y_test, (testing_samples, ))

training_samples = len(X_train)
testing_samples = len(X_test)

print('Number of training images: %i' % training_samples)
print('Number of testing images: %i' % testing_samples)
```

```python
lasso = linear_model.LogisticRegression(penalty='l1', multi_class='multinomial', solver='saga', tol=0.1)

t0 = time.time()

lasso.fit(X_train, y_train)

train_time = time.time() - t0
```

```python
num_classes = len(lasso.classes_)

sparsity = np.mean(lasso.coef_ == 0) * 100
score = lasso.score(X_test, np.reshape(y_test, (testing_samples, 1)))

print('Model trained in %.3f s' % train_time)
print('Sparsity with L1 regularization: %.2f' % sparsity)
print('Test score with L1 penalty: %.4f' % score)
```

```python
coef = lasso.coef_.copy()
plt.figure(figsize=(10, 5))
scale = np.abs(coef).max()
for i in range(num_classes):
    l1_plot = plt.subplot(2, 5, i + 1)
    l1_plot.imshow(coef[i].reshape(28, 28), interpolation='nearest', cmap=plt.cm.RdBu, vmin=-scale, vmax=scale)
    l1_plot.set_xticks(())
    l1_plot.set_yticks(())
    l1_plot.set_xlabel('Class %i' %i)
plt.suptitle('Classification vector for...')

plt.show()
```

### Plot ROC curves

```python
from itertools import cycle
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from scipy import interp

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

probs = lasso.decision_function(X_test)

for i in range(num_classes):
    y_true = (y_test == i).astype(int)
    fpr[i], tpr[i], _ = roc_curve(y_true, probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# Compute macro-average ROC curve and ROC area

# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))

# Then interpolate all ROC curves at this point
mean_tpr = np.zeros_like(all_fpr)
for i in range(num_classes):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i])

# Finally average it and compute AUC
mean_tpr /= num_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot all ROC curves
fig = plt.figure()

colors = cycle(['aqua', 'azure', 'lightblue', 'lightgreen', 'green', 'darkgreen', 'yellow', 'orange', 'red', 'plum'])
for i, color in zip(range(num_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label='ROC curve of class {0} (area = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot(fpr["macro"], tpr["macro"], label='macro-average ROC curve (area = {0:0.2f})'
        ''.format(roc_auc["macro"]),
    color ='navy', linestyle=':', linewidth=4)

plt.plot([0,1], [0,1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim(0.0, 1.05)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
fig.savefig('LASSO_ROC.png', bbox_inches='tight')
```
